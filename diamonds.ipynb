```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
# 使用 Raw 链接加载文件
csv_url = 'https://raw.githubusercontent.com/cliu001/diamonds/main/diamonds.csv'


# 加载文件到 DataFrame
df = pd.read_csv(csv_url, index_col=0)

# show the first 7 rows
print(df.head(7))

# Display the last 7 rows
print(df.tail(7))

#dataframe method
# List of numerical columns to consider
numerical_columns = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']

# Calculate descriptive statistics for the specified numerical columns
numerical_stats = df[numerical_columns].describe()

# Display the statistics
print(numerical_stats)

#series method
# Calculate value counts for the 'cut' column
cut_counts = df['cut'].value_counts()

# Calculate value counts for the 'color' column
color_counts = df['color'].value_counts()

# Calculate value counts for the 'clarity' column
clarity_counts = df['clarity'].value_counts()

# Display value counts for each categorical column
print("Cut Value Counts:\n", cut_counts)
print("\nColor Value Counts:\n", color_counts)
print("\nClarity Value Counts:\n", clarity_counts)


# Unique values in the 'cut' column
unique_cut_values = df['cut'].unique()

# Unique values in the 'color' column
unique_color_values = df['color'].unique()

# Unique values in the 'clarity' column
unique_clarity_values = df['clarity'].unique()

# Display unique values for each categorical column

print("Unique Cut Values:", unique_cut_values)
print("Unique Color Values:", unique_color_values)
print("Unique Clarity Values:", unique_clarity_values)


df.hist(figsize=(10, 8))
```
